```
import pandas as pd
import datetime as DT
import matplotlib.pyplot as plt


%matplotlib inline

train = pd.read_csv("train.csv")

train = pd.read_csv("train.csv", 
                  sep=r'[\s,]',              # 1
                  header=None, skiprows=1,
                  converters={               # 2
                      0: lambda x: DT.datetime.strptime(x, '%Y-%m-%d'),  
                      1: lambda x: DT.time(*map(int, x.split(':'))).isoformat()},
                  names=['Date', 'Time', 'Season', 'Holiday', 'Workingday', 'Weather', 'Temp', 'Atemp', 'Humidity', 'Windspeed', 'Casual', 'Registered', 'Count'])

train.head(2)
```

```
train.describe()
```

```
train.drop(['Casual', 'Registered'], axis=1, inplace=True)
train.describe()
```

```
plt.figure(figsize=(20,8))
plt.title("The new figure of rental regression prediction")
plt.plot(train["Date"], train["Count"])
plt.show()
```

```

plt.title("The time of count regression prediction")
plt.plot(train["Time"], train["Count"])
plt.show()
```
